```python
import pandas as pd
import numpy as np
```

```python
url_str = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url_str)
```

```python
df
```

```python
# Drop NaN records
df.dropna() # By default axis 0, which means Row-wise, inplace=True means we are applying changes to original record permanently
```

```python
df.dropna(axis=1) # axis 1, which means Column-wise - dropped columns with NaN values
```

```python
df.isnull().sum() # How many null records for each column
```

```python
# Dropthe rest, keep columns where we have more then 800 good records
df.dropna(axis=1, thresh=800)
```

```python
df.fillna(value="manu") # 
```

```python
df.fillna(value=df['Age'].mean()) # Fill NaN values with average age
```

```python
df.groupby('Survived').describe()['Fare']
```

```python
# Rotate the above data, means put columns as rows and vice-versa
df.groupby('Survived').describe()['Fare'].T # both same
df.groupby('Survived').describe()['Fare'].transpose()
```
