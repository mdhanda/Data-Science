```python
df1 = sns.load_dataset('tips') # Available in sns library by default
df1
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, style='smoker')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, hue = 'smoker', style='size')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, hue = 'smoker', style='time')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, size='size', hue = 'smoker', style='time')
```

```python
sns.relplot(kind = 'line', x= 'total_bill', y = 'tip', data=df1)
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, col='time') # Based on 'time' column, it separated graphs
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, col='size')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, col='day', hue = 'smoker', style='time')
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data=df1, col='sex') # Who's giving more tips.. male or female
```

```python
sns.catplot(x= 'day', y = 'total_bill', data=df1, hue='time')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data=df1, kind='violin')
```

```python
sns.catplot(x= 'day', y = 'total_bill', data=df1, kind='swarm')
```
