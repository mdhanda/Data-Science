## Handling Missing Values

```python
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0]
for feature in feature_nan:
    print("The feature is {} and missing values in {}%".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

```python
## Replacing Categorical Features
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes=='O']
categorical_features
```

```python
# First technique (Better way)
# Replace the missing values with our own category named 'Missing'
data = dataset.copy()
def replace_cat_features(data, categorical_features):
    data[categorical_features] = data[categorical_features].fillna('Missing')
    return data
```

```python
replace_cat_features(data, categorical_features)
```

```python
# Second technique
# For Categorical Values, Replace the missing values with 'mode'
data = dataset.copy()
def replace_cat_features(data, categorical_features):
    data[categorical_features] = data[categorical_features].fillna(data[categorical_features].mode())
    return data
```

```python
replace_cat_features(data, categorical_features)
```

```python
## Handling Numerical Features with NaN
numerical_features_nan = [feature for feature in dataset.columns 
                          if dataset[feature].dtypes != 'O' and dataset[feature].isnull().sum() > 0]
numerical_features_nan
```

```python
data = dataset.copy()
for feature in numerical_features_nan:
    ## since there are outliers in numerical features listed above, we are going to replace with median
    median_values=data[feature].median()
    ## create a new feature to capture NaN values
    data[feature+'NaN']=np.where(data[feature].isnull(), 1, 0)
    data[feature].fillna(median_values, inplace=True)
```

```python
data
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O'] # 'O' is same as 'object'
numerical_features
```

```python
## Handling Temporal Variables (datet time Variables)
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature] # or 'Mo' in feature]
print(year_feature)
```

```python
data = dataset.copy()

# We are particularly interested in Age of House
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    data[feature+'AgeOfHouse']=data['YrSold'] - data[feature]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualize all the columns in the dataframe
pd.set_option('display.max_columns', None)
```

```python
# Problem Stmt: Predict the price of Tkt
# Feature Engineering -> Make this data meaningful
```

```python
# Date of Journey -> Convert to numbers
```

```python
dataset['Date']= dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month']= dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year']= dataset['Date_of_Journey'].str.split('/').str[2]

## OR We can use lambda functions for the same
dataset['Date']= dataset['Date_of_Journey'].apply(lambda x: x.split('/')[0])
dataset['Month']= dataset['Date_of_Journey'].apply(lambda x: x.split('/')[1])
dataset['Year']= dataset['Date_of_Journey'].apply(lambda x: x.split('/')[2])
```

```python
dataset
```

```python
dataset.info()
```

```python
## As of now Date, Month, Year are of type object, let's convert them to int
dataset['Date']=dataset['Date'].astype(int)
dataset['Month']=dataset['Month'].astype(int)
dataset['Year']=dataset['Year'].astype(int)
dataset.info()
```

```python
# We don't need Data_of_Journey anymore, let's just drop it
dataset.drop('Date_of_Journey', axis=1, inplace=True)
dataset.head(2)
```

```python
# As we have Duration, we don't need Dep_Time, Arrival_Time.
# But, just in case we don't have Duration, How can we handle Arrival_Time formatting as given in "01:10 22 Mar"
# Lets get rid of ' 22 Mar' from 'Arrival_Time'
dataset['Arrival_Time']=dataset['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
# Let's CreateHour, Min columns for Arrival and Departure Times
dataset['Arrival_Hour']=dataset['Arrival_Time'].apply(lambda x: x.split(':')[0])
dataset['Arrival_Min']=dataset['Arrival_Time'].apply(lambda x: x.split(':')[1])

dataset['Dep_Hour']=dataset['Dep_Time'].apply(lambda x: x.split(':')[0])
dataset['Dep_Min']=dataset['Dep_Time'].apply(lambda x: x.split(':')[1])
```

```python
dataset.head(2)
```

```python
dataset['Arrival_Hour']=dataset['Arrival_Hour'].astype(int)
dataset['Arrival_Min']=dataset['Arrival_Min'].astype(int)

dataset['Dep_Hour']=dataset['Dep_Hour'].astype(int)
dataset['Dep_Min']=dataset['Dep_Min'].astype(int)

dataset.drop('Arrival_Time', axis=1, inplace=True)
dataset.drop('Dep_Time', axis=1, inplace=True)
dataset.head(2)
```

```python
# Lets convert Duration in Minutes
import re

def HourStrToMins(HrMinStr):
    if(type(HrMinStr) is int):
        return HrMinStr
    
    mins = 0
    hm = re.findall("\d+", HrMinStr) #returns list containing ['hr', 'min']
    if('h' in HrMinStr):
        hm = [int(x) for x in hm]
        mins += int(hm[0])*60
    elif('m' in HrMinStr):
        mins += int(hm[0])
    return mins
    
dataset['Duration']=dataset['Duration'].apply(HourStrToMins)
dataset.head(2)
```

```python
# Lets deal with Total_Stops now
dataset['Total_Stops'].unique()
```

```python
dataset['Total_Stops'].mode() # To replace nan, let's find mode
```

```python
dataset['Total_Stops'] = dataset['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```
