```python
# Object Oriented Programming System (Oops)
class Phone:
    def set_color(self, color):
        self.color = color
        
    def set_cost(self, cost):
        self.cost = cost
        
    def show_color(self):
        return self.color
    
    def show_cost(self):
        return self.cost
    
    def make_call(self):
        print("making call")
        
    def play_game(self):
        print("playing game")
```

```python
p1 = Phone()
```

```python
p1.make_call()
```

```python
p1.play_game()
```

```python
p1.set_cost(200)
p1.set_color("black")
```

```python
p1.show_cost()
```

```python
class Employee():
    
    def __init__(self, name, age, salary, gender):
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
        
    def employee_details(self):
        print("Employee Name is ", self.name)
        print("Employee Age is ", self.age)
        print("Employee Salary is ", self.salary)
        print("Employee Gender is ", self.gender)
        
emp = Employee("Manu", 23, 85400, "Male")
emp.employee_details()
        
```

```python
class Vehicle():
    
    def __init__(self, mileage, cost):
        self.mileage = mileage
        self.cost = cost
        
    def show_details(self):
        print("Vehicle mileage is ", self.mileage)
        print("Vehicle cost is ", self.cost)
        
v1 = Vehicle(35000, 2400)
v1.show_details()
```

```python
# Types of Inheritance
# Single, Multiple, Multi-level, Hybrid

class Car(Vehicle):
    
    def show_car(self):
        print("i am a car")
        
c1 = Car(48500, 4500)
c1.show_details()
c1.show_car()
```

```python
class Car(Vehicle):
    
    def __init__(self, mileage, cost, tyres, hp):
        super().__init__(mileage, cost)
        self.tyres = tyres
        self.hp = hp
        
    def show_car(self):
        print("Hello, I'm a car with mileage {}, costing ${}. I've {} tyres and power of {} hp".format(self.mileage, self.cost, self.tyres, self.hp))
        
c1 = Car(48500, 4500, 4, 1200)
c1.show_details()
c1.show_car()
```

```python
# Multiple Inheritance

class Parent1():
    
    def __init__(self):
        print("In Parent1")
        
    def assign_string_one(self, str1):
        self.str1 = str1
        
    def show_string_one(self):
        return self.str1

class Parent2():
    
    def __init__(self):
        print("In Parent2")

    def assign_string_two(self, str2):
        self.str2 = str2
        
    def show_string_two(self):
        return self.str2

class Derived(Parent1, Parent2):
    
    def __init__(self):
        print("Derived from Multiple parents")
        
    def assign_string_three(self, str3):
        self.str3 = str3
        
    def show_string_three(self):
        return self.str3
    
    
d = Derived()
d.assign_string_one("One")
d.assign_string_two("Two")
d.assign_string_three("Three")
```

```python
d.show_string_one()
```

```python
d.show_string_two()
```

```python
d.show_string_three()
```

```python
# Multi-level Inheritance [Parent-Child-GrandChild]

class Parent():
    
    def set_name(self, name):
        self.name = name
        
    def show_name(self):
        return self.name
    
class Child(Parent):
    
    def set_age(self, age):
        self.age = age
        
    def show_age(self):
        return self.age
    
class GrandChild(Child):
    
    def set_gender(self, gender):
        self.gender = gender
        
    def show_gender(self):
        return self.gender
```

```python
gc = GrandChild()
gc.set_gender("Male")
gc.show_gender()
```
